```python
image_dir = "/content/darknet/data/images"
train_txt = "/content/darknet/data/train.txt"
test_txt = "/content/darknet/data/test.txt"
```

```python
import random
import os

# Collect all image paths
all_images = [os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.lower().endswith(".jpg")]
random.shuffle(all_images)
```

```python
# 80-20 split
train_count = int(0.8 * len(all_images))
train_images = all_images[:train_count]
test_images = all_images[train_count:]
```

```python
#  Write to train.txt and test.txt
with open(train_txt, 'w') as f:
    f.write('\n'.join(train_images))
```

```python
with open(test_txt, 'w') as f:
    f.write('\n'.join(test_images))
```

```python
print(f"Train images: {len(train_images)}, Test images: {len(test_images)}")
```
